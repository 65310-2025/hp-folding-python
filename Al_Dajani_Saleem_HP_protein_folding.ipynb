```python
# STEP 1: Imports
import sys
sys.path.append(".")

import configparser
import matplotlib.pyplot as plt
import imageio
import os
from protein_class import Protein
from utils import Configuration
from plots import view
from IPython.display import Image, display
from zipfile import ZipFile

# STEP 2: Define HP sequences for A–Z
alphabet_sequences = {
    'A': 'HPPHPHPPH',
    'B': 'HPHPHPHPH',
    'C': 'HPPHHPPH',
    'D': 'HPHPPHPH',
    'E': 'HHPPPHPPH',
    'F': 'HHPHHPH',
    'G': 'HPPHPHPHH',
    'H': 'HPHPHPH',
    'I': 'HHPPHH',
    'J': 'PPHHPPHH',
    'K': 'HPHHPHPH',
    'L': 'HHPPHHPP',
    'M': 'HHPHPHPH',
    'N': 'HPHPHPPH',
    'O': 'HPHPHPHPH',
    'P': 'HHPPHHPH',
    'Q': 'HPHPPHHP',
    'R': 'HPHPHHPH',
    'S': 'HPPHPHPH',
    'T': 'HPHPPPHPH',
    'U': 'HHPPHPPHH',
    'V': 'HPHPHPHP',
    'W': 'HPHPHPPHPH',
    'X': 'HPHHPHPHH',
    'Y': 'HHPHPHPPH',
    'Z': 'HPHPHPPHP'
}

# STEP 3: Prepare output folders
os.makedirs("alphabet_png", exist_ok=True)
os.makedirs("alphabet_gif", exist_ok=True)
os.makedirs("alphabet_frames", exist_ok=True)

# STEP 4: Loop through letters A–Z
parser = configparser.ConfigParser()
parser.read("config_test.txt")

for letter, seq in alphabet_sequences.items():
    print(f"Processing {letter}: {seq}")
    config = Configuration(parser)
    config.seq = seq
    config.use_struct = False
    config.gif = True

    protein = Protein(config)
    protein.evolution()

    # --- Save final fold as PNG
    fig, ax = plt.subplots()
    view(protein, save=False, tit=f"Letter: {letter}")
    png_path = f"alphabet_png/{letter}.png"
    fig.savefig(png_path, dpi=150, bbox_inches="tight")
    plt.close(fig)

    # --- Save folding GIF
    frames = []
    for i, struct in enumerate(protein.gif_struct):
        fig, ax = plt.subplots()
        x, y = zip(*struct)
        ax.plot(x, y, alpha=0.5)
        for j, coord in enumerate(struct):
            ax.scatter(x[j], y[j], marker='$'+protein.seq[j]+'$', s=30, color='red')
        ax.set_xlim(min(x)-6, max(x)+6)
        ax.set_ylim(min(y)-6, max(y)+6)
        ax.axis('off')
        fname = f"alphabet_frames/{letter}_{i:03d}.png"
        plt.savefig(fname, dpi=100, bbox_inches='tight')
        frames.append(fname)
        plt.close()

    gif_path = f"alphabet_gif/{letter}.gif"
    with imageio.get_writer(gif_path, mode='I', duration=0.1) as writer:
        for f in frames:
            writer.append_data(imageio.imread(f))

    # Cleanup frames
    for f in frames:
        os.remove(f)

# STEP 5: Zip all outputs
!zip -r protein_alphabet_outputs.zip alphabet_png alphabet_gif

# STEP 6: Show a sample (first 5 letters)
from IPython.display import display, HTML
html = ""
for letter in list(alphabet_sequences.keys())[:5]:
    html += f"<div style='display:inline-block;margin:10px;text-align:center'><img src='alphabet_gif/{letter}.gif' width='120'><br>{letter}</div>"
display(HTML(html))

print("✅ All done! Download everything below:")
```
